## Interpretação dos Testes Estatísticos

Foram realizados testes Qui-Quadrado de independência para verificar a existência de associação estatística entre as variáveis analisadas.

### Região × Gravidade

O teste rejeitou a hipótese nula de independência (p < 0.05), indicando que a distribuição da gravidade dos acidentes difere significativamente entre as regiões brasileiras.

### Tipo de Pista × Gravidade

O teste também rejeitou a hipótese nula de independência (p < 0.05), indicando que a gravidade dos acidentes está associada ao tipo de pista em que ocorreram.

Portanto, os resultados sugerem que tanto fatores regionais quanto características da infraestrutura viária estão relacionados à severidade dos sinistros.
